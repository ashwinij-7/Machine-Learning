Import Dataset

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("purusinghvi/email-spam-classification-dataset")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

file_path = os.path.join(path)

print("Dataset directory:", file_path)
```

```python
print(os.listdir(path))
```

```python
csv_path = os.path.join(path, 'combined_data.csv')
df = pd.read_csv(csv_path)

print(df.head())
```

Model Training

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
X = df['text']
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vectorized, y_train)
```

Evaluation

```python
y_pred = model.predict(X_test_vectorized)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
train_pred = model.predict(X_train_vectorized)

# Testing predictions
test_pred = model.predict(X_test_vectorized)

# Accuracy
train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```
